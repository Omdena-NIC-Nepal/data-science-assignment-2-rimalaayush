## Task 7: Merging and Joining DataFrames
- Merge two DataFrames on a common column.
- Join two DataFrames using their indices.

```python
# Sample DataFrame 1
named_data = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
}
name_df = pd.DataFrame(named_data)

# Sample DataFrame 2
age_data = {
    'ID': [3, 4, 5, 6],
    'Age': [25, 30, 35, 40]
}
age_df = pd.DataFrame(age_data)
merged_df = pd.merge(name_df, age_df, on='ID', how='outer')
merged_df
```

## Task 8: Working with Dates and Times
- Create a datetime index for the DataFrame.
- Convert a column to datetime and extract date components.

```python

sale_data = {
    'Date String': ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01'],
    'Total Sales': [10, 20, 30, 40]
}
df = pd.DataFrame(sale_data)
df['Date Formatted'] = pd.to_datetime(df['Date String'])

df.set_index('Date Formatted', inplace=True)
# Extract date components
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['DayOfWeek'] = df.index.dayofweek
df['WeekOfYear'] = df.index.isocalendar().week
df

```
